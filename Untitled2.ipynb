```python
class Class1:
  def test_class1(self):
    print("this is class1")

class Class2(Class1):
  def test_class2(self):
    print("this is class2")

class Class3(Class2):
  def test_class3(self):
    print("this is class3")
```

```python
obj = Class3()
```

```python
obj.test_class1()
```

```python
obj.test_class2()
```

```python
class Class1:
  def test_class1(self):
    print("this is class1")

class Class2(Class1):
  def test_class2(self):
    print("this is class2")

class Class3(Class2,Class3):
  def test_class3(self):
    print("this is class3")
```

```python
obj = Class3()
```

```python
obj.test_class1()
```

```python
class data_science:
    def syllabus(self):
        print("this is my syllabus for data science")

class web_dev:
    def syllabus(self):
        print("this is my syllabus for web dev")
```

```python
ds = data_science()
wd = web_dev()
```

```python
def class_parcer(class_obj):
    for i in class_obj:
        i.syllabus()
```

```python
class_obj = [data_science, web_dev]
```

```python
class_parcer(class_obj)
```
