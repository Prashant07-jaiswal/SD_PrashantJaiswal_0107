```python
l= [1,2,3,4,"sudh","kumar",324,34.456,"abd"]
intl=[]
strl=[]
for i in l:
  if type(i)==int:
    intl.append(i)
  elif type(i)==str:
    strl.append(i)
print(intl)
print(strl)
```

```python
l2 = [20,30,28.34,87.32,"hello","ravi",69.69,True,False]
num = []
fl =[]
string = []
booll = []
for i in l2:
  if type(i)==int:
    num.append(i)
  elif type(i)==float:
    fl.append(i)
  elif type(i)==bool:
    booll.append(i)
  else:
    string.append(i)
print(num)
print(fl)
print(string)
print(booll)
```

```python
x = int(input("Enter a number: "))
temp = x
reverse = 0
while x>0:
  digit = x%10
  reverse = (reverse*10)+digit
  x = x//10

if reverse == temp:
  print(temp, "is a palindrome number")
else:
  print(temp, "is not a palindrome number")
```
